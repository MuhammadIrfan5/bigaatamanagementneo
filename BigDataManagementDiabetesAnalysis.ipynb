```python
df = df.dropna()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters)

kmeans.fit(df)

cluster_labels = kmeans.labels_
df['Cluster'] = cluster_labels
print(cluster_labels)

plt.scatter(df['HighBP'], df['Diabetes_binary'], c=cluster_labels)
plt.xlabel("BP")
plt.ylabel("Diabetes_binary")
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
independent_vars = df[['HighBP', 'HighChol', 'BMI', 'HeartDiseaseorAttack', 'GenHlth', 'PhysHlth', 'DiffWalk', 'Age']]
dependent_variable = df['Diabetes_binary']

decision_tree = DecisionTreeClassifier()

decision_tree.fit(independent_vars, dependent_variable)

predicted_labels = decision_tree.predict(independent_vars)


print(predicted_labels)
confusion_matrix = pd.crosstab(dependent_variable, predicted_labels, rownames=['Actual'], colnames=['Predicted'])

print(confusion_matrix)

accuracy = decision_tree.score(independent_vars, dependent_variable)

print(accuracy)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(decision_tree, out_file='decision_tree.dot', feature_names=independent_vars.columns, class_names=['0', '1'], label='all', filled=True, rounded=True, special_characters=True)

!apt-get install graphviz -y

!dot -Tpng decision_tree.dot -o decision_tree.png

from IPython.display import Image
Image(filename='decision_tree.png')
```
